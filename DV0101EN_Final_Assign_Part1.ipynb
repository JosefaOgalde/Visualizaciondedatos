####  TASK 1.2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


<details><summary>Click here for a hint</summary>

<p>
    You will require:-
    <br>create a separate dataframe where the column recession has a value of '1'
    <br>to group the year, vehicle_type and calculate the average on the 'Automobile Sales'
    <br>one way is to -
    <br>use as_index as false else you will endup with multiple-indexed datafame

    <br>make use of .plot() with kind = 'line'
    <br>do not forget to include labels and title

````python
# prompt: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?

# Create a DataFrame for recession periods only
recession_df = df[df['Recession'] == 1]

# Group data by year, vehicle type, and calculate the mean automobile sales
sales_by_vehicle_type = recession_df.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Set 'Year' as the index and pivot the table for plotting
sales_by_vehicle_type = sales_by_vehicle_type.set_index('Year')
sales_pivot = sales_by_vehicle_type.pivot(columns='Vehicle_Type', values='Automobile_Sales')

# Create a line plot to visualize sales trends for different vehicle types
plt.figure(figsize=(12, 6))
for col in sales_pivot.columns:
    plt.plot(sales_pivot.index, sales_pivot[col], label=col, marker='o', linestyle='-')

plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Sales Trend of Different Vehicle Types during Recession')
plt.xticks(sales_pivot.index, rotation=45, ha='right')
plt.legend()
plt.grid(True)
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.

<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
# prompt: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.

import matplotlib.pyplot as plt

# Create a figure and axes for subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Filter data for recession periods
recession_data = df[df['Recession'] == 1]

# Plot GDP variation during recession
axes[0].plot(recession_data['Year'], recession_data['GDP'], marker='o', linestyle='-')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('GDP')
axes[0].set_title('GDP Variation During Recession')
axes[0].grid(True)


# Filter data for non-recession periods
non_recession_data = df[df['Recession'] == 0]

# Plot GDP variation during non-recession
axes[1].plot(non_recession_data['Year'], non_recession_data['GDP'], marker='o', linestyle='-')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('GDP')
axes[1].set_title('GDP Variation During Non-Recession')
axes[1].grid(True)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Display the plots
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
# prompt: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.

# Assuming 'df' is already defined from the previous code

# Filter data for non-recession periods
non_recession_data = df[df['Recession'] == 0]

# Create the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(non_recession_data['Month'], non_recession_data['Automobile_Sales'], s=non_recession_data['Seasonality_Weight'] * 100, alpha=0.6)  # Adjust the size multiplier (100) as needed
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Seasonality impact on Automobile Sales')
plt.grid(True)
plt.show()
````

````python
# prompt: How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend
# Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble
# Title this plot as 'Seasonality impact on Automobile Sales'

# Assuming 'df' is already defined from the previous code

# Filter data for non-recession years and group by month
non_recession_df = df[df['Recession'] == 0]
monthly_sales = non_recession_df.groupby('Month')['Automobile_Sales'].mean().reset_index()

# Merge with seasonality weights
monthly_sales = pd.merge(monthly_sales, non_recession_df[['Month', 'Seasonality_Weight']], on='Month', how='left').drop_duplicates(subset=['Month'])

# Create the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(monthly_sales['Month'], monthly_sales['Automobile_Sales'],
            s=monthly_sales['Seasonality_Weight'] * 500, # Adjust size multiplier as needed
            alpha=0.6)

# Customize the plot
plt.xlabel('Month')
plt.ylabel('Average Automobile Sales')
plt.title('Seasonality impact on Automobile Sales')
plt.grid(True)
plt.show()
````
